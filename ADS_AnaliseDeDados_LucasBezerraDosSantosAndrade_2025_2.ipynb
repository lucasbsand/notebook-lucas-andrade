## Exercício 2 - Carregando os dados do estudo de caso com o *Jupyter* e o pandas.

```python
import pandas as pd

df = pd.read_excel('/content/default_of_credit_card_clients__courseware_version_1_21_19.xls', na_values=['Not available'])

df.head()
```

### 1. Quantas colunas os dados contêm?

```python
print(f"O arquivo contêm {df.shape[1]} colunas.")
```

### 2. Quantas linhas (amostras) os dados contêm?

```python
print(f"O arquivo contêm {df.shape[0]} linhas.")
```

### 3. Que tipos de características existem. Quais são categóricas e quais são numéricas?

```python

categorical_cols = ['SEX', 'EDUCATION', 'MARRIAGE',
                    'PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',
                    'default payment next month']

numeric_cols = ['LIMIT_BAL', 'AGE',
                'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3',
                'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
                'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
                'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

print(f"Colunas categóricas: {categorical_cols}\n")
print(f"Colunas numéricas: {numeric_cols}")
```

### 4. Qual é a aparência dos dados segundo essas características?
Informações colunas numéricas:


```python
df[numeric_cols].describe()
```

Informações colunas categóricas:

```python
for col in categorical_cols:
  print(f"\nFrequência de valores na coluna:")
  print(df[col].value_counts())
```

### 5. Existem dados faltando?

```python
all_cols = categorical_cols + numeric_cols

for col in all_cols:
  print(f"Faltam {df[col].isnull().sum()} dados na coluna {col}.\n")
```
